```python
import keras
from keras import backend as k
from keras.models import load_model
print('done')
```

```python
pm=load_model('myanprproject.h5')
print(pm.summary())
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
path='./Predict'
p=os.listdir(path)
print(p)
ch='0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
for i in range(0,len(p)):
    op=path+'/'+p[i]
    img=cv2.imread(op)
    plt.subplot(3,4,i+1)
    plt.imshow(img)
    img=img.reshape(1,28,28,3)
    output=pm.predict_classes(img)[0]
    plt.title('Predicted: '+ch[output])
    plt.axis('off')
    
```
